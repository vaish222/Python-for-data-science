```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig  =  plt.figure()  #create  a figure object 

ax = fig.add_axes([0,0,1,1]) #added axes to figure canvas
ax.plot(x,y) #plotting x and y 
#set the labels and titles to match the plot
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
#create a figure object and put two  axes located at [0,0,1,1] and [0.2,.5,.2,.2]
fig  = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,.5,.2,.2])

ax1.plot(x,y,color='red')
ax2.plot(x,y,color='red')
ax1.set_xlabel('x')
ax2.set_ylabel('y')
```

```python
#create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]. Also, use x,y,z to recreate the plots. 
#set the limits to x and y plot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])

ax.plot(x,z)
ax.set_xlabel('x')
ax.set_ylabel('z')


#inside plot
ax2.plot(x,y)
ax2.set_title('zoom')
ax2.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
#subplot to create a plot
fig,axes = plt.subplots(1,2,figsize=(12,2))

axes[0].plot(x,y, color='blue',ls='--',lw=3)
axes[1].plot(x,z,color='red',lw=3)
```

```python
plt.show()
```
